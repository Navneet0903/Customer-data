```python
# Customer Segmentation Using K-Means
# Author: Navneet Raj

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

# Load the dataset
df = pd.read_csv("customer_data.csv")  # Make sure this file is uploaded in the same directory
df.head()
```

```python
# Select relevant features for clustering
features = ['Age', 'Annual Income (k$)', 'Frequency', 'Spending Score (1–100)']
X = df[features]
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

```python
# Elbow method to determine optimal number of clusters
inertia = []
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)
```

```python
# Plot the Elbow curve
plt.figure(figsize=(8, 4))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
# Apply KMeans with k=4 (or use the optimal k based on elbow)
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1–100)', hue='Cluster', palette='Set2')
plt.title('Customer Segmentation (2D)')
plt.show()
```

```python
# 3D visualization of clusters
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(df['Age'], df['Annual Income (k$)'], df['Spending Score (1–100)'],
                c=df['Cluster'], cmap='rainbow', s=60)
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')
plt.title('Customer Segmentation (3D)')
plt.show()
```
